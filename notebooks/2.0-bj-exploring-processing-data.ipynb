```python
# imports
import pandas as pd
import numpy as np
import os
```

```python
raw_data_path = os.path.join(os.path.pardir,'data','raw')
print(raw_data_path)
train_file_path = os.path.join(raw_data_path,'train.csv')
test_file_path = os.path.join(raw_data_path,'test.csv')
```

```python
train_df=pd.read_csv(train_file_path, index_col='PassengerId')
test_df=pd.read_csv(test_file_path, index_col='PassengerId')
```

# Summary Statistics

```python
# describe can be used to get statistical information about all columns within a data frame
df.describe()
```

```python
# numerical feature
# centrality measures

print('Mean Fare: {0}'.format(df.Fare.mean())) # mean
print('Median Fare: {0}'.format(df.Fare.median())) # median
```

```python
# dispersion measures

print('Min Fare: {0}'.format(df.Fare.min()))
print('Max Fare: {0}'.format(df.Fare.max()))
print('Fare range: {0}'.format(df.Fare.max()-df.Fare.min()))
print('25 percentile: {0}'.format(df.Fare.quantile(.25)))
print('50 percentile: {0}'.format(df.Fare.quantile(.5)))
print('75 percentile: {0}'.format(df.Fare.quantile(.75)))
print('Fare Variance: {0}'.format(df.Fare.var()))
print('Fare standard deviation: {0}'.format(df.Fare.std()))
```
