# Grouping and aggregation function

```python
df.groupby('Sex').Age.mean()
```

```python
df.groupby(['Pclass']).Fare.median()
```

```python
df.groupby(['Pclass']).Age.median()
```

```python
df.groupby(['Pclass'])['Fare','Age'].median()
```

```python
df.groupby(['Pclass']).agg({'Fare':'mean','Age':'median'})
```

```python
# A complicated aggregation gets called later on
aggregation={# work on just Fare column first
    'Fare':{
        'mean_Fare':'mean',
        'median_Fare':'median',
        'max_Fare':max,
        'min_Fare':np.min,
    },
    'Age':{# work on the Age column
        'mean_age':'mean',
        'median_age':'median',
        'min_age':min,
        'max_age':max,
        'age_range':lambda x: max(x)-min(x) # calculate the range for each group using lambda function 
    }
}
```

```python
df.groupby(['Pclass']).agg(aggregation)
```

```python
# multi column grouping
df.groupby(['Pclass','Embarked']).Fare.median()
```
